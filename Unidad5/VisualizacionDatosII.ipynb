```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_canada = pd.read_excel(
    './Data/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows = range(20),
    skipfooter = 2
)
```

```python
df_canada.head()
```

## Graficoos de dispersion (Scatter)

```python
years=list(map(int,range(1980,2014)))
df_tot = pd.DataFrame(df_canada[years].sum(axis=0))
df_tot.head(5)
df_tot.index = map(int,df_tot.index)
df_tot.reset_index(inplace=True)
df_tot.columns = ['Año','Total']
```

```python
df_tot.head()
```

```python

df_tot.plot(kind='scatter',x = 'Año', y = 'Total',figsize=(10,6),
           color='darkblue',
           title='Total de Inmigracion a Canada de 1908 al 2013')
plt.xlabel('Año')
plt.xlabel('Numero de inmigrandtes')
plt.show()
```

```python

df_tot.plot(x = 'Año', y = 'Total',figsize=(10,6),
           color='darkblue',
           title='Total de Inmigracion a Canada de 1908 al 2013')
plt.xlabel('Año')
plt.xlabel('Numero de inmigrandtes')
plt.show()
```

#### Obtener la line de mejor ajuste para la linea de regresion

x = coordenadas x de los datos <br>
y = coordenadas y de los datos <br>
deg = grado de ajuste del polinomio donde 1 es lineal, 2 es cuadratico

```python
x = df_tot['Año']
y = df_tot['Total']
fit = np.polyfit(x,y, deg =1)
```

```python

df_tot.plot(kind='scatter',x = 'Año', y = 'Total',figsize=(10,6),
           color='darkblue',
           title='Total de Inmigracion a Canada de 1908 al 2013')
plt.xlabel('Año')
plt.xlabel('Numero de inmigrandtes')
plt.plot(x, fit[0]*x+fit[1],color='red')
plt.annotate('y = {0:.0f} x + {1:.0f}'.format(fit[0],fit[1]),xy=(2000,150000))
plt.show()
```

#### Graficos de Burbujas

```python
df_canada_t = df_canada.set_index('OdName',inplace=True)
df_canada_t = df_canada[years].transpose()
df_canada_t.index = map(int, df_canada_t.index)
df_canada_t.index.name = 'Year'
df_canada_t.reset_index(inplace = True)
df_canada_t.head()
```

```python
df_canada_t.head()
```

```python
#Proceso de normalizacion de los datos usando la tecnica de escalmiento de
#caracteristicas
norm_mexico = (df_canada_t['Mexico'] - df_canada_t['Mexico'].min())/ (df_canada_t['Mexico'].max()-df_canada_t['Mexico'].min())
norm_argentina = (df_canada_t['Argentina'] - df_canada_t['Argentina'].min())/ (df_canada_t['Argentina'].max()-df_canada_t['Argentina'].min())
norm_brasil = (df_canada_t['Brazil'] - df_canada_t['Brazil'].min())/ (df_canada_t['Brazil'].max()-df_canada_t['Brazil'].min())
```

```python
norm_mexico.head()
```

```python
ax0 = df_canada_t.plot(
    kind='scatter',
    x = 'Year', y = 'Mexico',
    figsize = (10,6),
    alpha = 0.5, color = 'green',
    s = norm_mexico * 2000 +10,
    xlim = (1975,2015)
)
ax1 = df_canada_t.plot(
    kind='scatter',
    x = 'Year', y = 'Argentina',
    figsize = (14,8),
    alpha = 0.5, color = 'blue',
    s = norm_argentina * 2000 +10,
    ax = ax0
)

ax2 = df_canada_t.plot(
    kind='scatter',
    x = 'Year', y = 'Brazil',
    figsize = (14,8),
    alpha = 0.5, color = 'red',
    s = norm_brasil * 2000 +10,
    ax = ax0
)

ax0.set_title('Inmigracion de Mexico, Argentina y Brasil de 1980 a 2013')
ax0.set_ylabel('Numero de inmigrantes')

ax0.legend(['Mexico','Argentina'],loc='upper left',fontsize='x-large')
```
